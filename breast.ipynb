```python
#abrir o arquivo
import pandas as pd
dados = pd.read_csv("/content/asd.csv", sep=',')
dados = dados.drop(dados[dados.BareNuclei == '?'].index)

print(dados.head)
```

```python
#separar atributos e classes
dados_atributos = dados.drop(columns = ['Class'])
output = dados['Class']
```

```python
#verificar a frequencia dos dados
from collections import Counter
atributos_count = Counter(output)
print(atributos_count)
```

```python
#balancear as classes
from imblearn.over_sampling import SMOTE
#construir um onketo a partir da classe SMOTE
resampler = SMOTE()
#balancear os dados

dados_atributos_b, output_b = resampler.fit_resample(dados_atributos, output)
#verificar a distribuição das classes após o balanceamento
output_b_count = Counter(output_b)
print(output_b_count)
```

```python
#tipos de objetos resultantes do SMOTE
print(type(output_b_count))
```

```python
#segmentar a base em dados para treinamento e dados para testes
#HOLD OUT (70/30)
from sklearn.model_selection import train_test_split
atr_train, atr_test, class_train, class_teste = train_test_split(dados_atributos_b, output_b, test_size = 0.3)
```

```python
#treinar uma arvore de decisao
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
breast_tree = tree.fit(atr_train, class_train)
```
